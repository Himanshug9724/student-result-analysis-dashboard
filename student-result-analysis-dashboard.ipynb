```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_excel(r"C:\Users\buyt2\python\student_marks.xlsx")


# Subject-wise average
subject_avg = df.mean(numeric_only=True)

# Top performers
top_students = df.sort_values("Total", ascending=False).head(5)

# Visualization
subject_avg.plot(kind='bar', title="Subject-wise Average Marks")
plt.show()

plt.pie(df['Math'].value_counts(), labels=df['Math'].value_counts().index,
        autopct='%1.1f%%')
plt.title("Math Score Distribution")
plt.show()
```
